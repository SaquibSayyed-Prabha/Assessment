```python
# Create a Student DataFrame containing: 
# - Missing values in Age and Marks columns. 
# - One duplicate student record. 

# Perform the following operations: 
# 1. Find missing values. 
# 2. Count missing values. 
# 3. Replace missing values with 0. 
# 4. Replace missing age with average age. 
# 5. Identify duplicate records. 
# 6. Remove duplicate records. 
```

```python
# Create a Student DataFrame containing: 
# - Missing values in Age and Marks columns. 
# - One duplicate student record. 
import pandas as pd

data = {
    "Name" : ["Saquib", "Saquib", "Suraj", "Atharva", "Jitendra"],
    "Age" : [17, 17, 18, None, 20],
    "City" : ["Katraj", "Katraj", "Nashik", "Narhe", "Rajasthan"],
    "Marks" : [89, 89, None, 87, 86]
}

df = pd.DataFrame(data)
print(df)
```

```python
# 1. Find missing values. 
print(df.isnull())
```

```python
# 2. Count missing values.
print(df.isnull().sum())
```

```python
# 4. Replace missing age with average age. 
avg_age = df["Age"].mean()
df["Age"] = df["Age"].fillna(avg_age)

print(df)
```

```python
# 3. Replace missing values with 0. 
df = df.fillna(0)
print(df)
```

```python
# 5. Identify duplicate records. 
print(df[df.duplicated()])
```

```python
# 6. Remove duplicate records. 
df = df.drop_duplicates()

print(df)
```
