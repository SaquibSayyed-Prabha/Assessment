```python
# Create a Pandas DataFrame for 5 students containing: 
# - Name 
# - Age 
# - City 
# - Marks 

# 1. Create the DataFrame. 
import pandas as pd

data = {
    "Name" : ["Saquib", "Suraj", "Atharva", "Jitendra"],
    "Age" : [17, 18, 18, 20],
    "City" : ["Katraj", "Nashik", "Narhe", "Rajasthan"],
    "Marks" : [89, 90, 87, 86]
}

df = pd.DataFrame(data)
```

```python
# 2. Display the DataFrame. 
print(df)
```

```python
# 3. Save the DataFrame into a CSV file named students.csv
df.to_csv("students.csv", index = False)
```

```python
# 4. Read the CSV file again. 
df = pd.read_csv("students.csv")
```

```python
# 5. Display the loaded data.
print(df)
```
