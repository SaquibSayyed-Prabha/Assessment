```python
# Create an Employee DataFrame containing: 
# - Employee ID 
# - Name 
# - Department 
# - Salary 
# - City 

import pandas as pd

Employee = {
    "Emp ID" : [101, 102, 103, 104, 105, 106], 
    "Name" : ["Saquib", "Atharva", "Jitendra", "Aaryan", "Hemant", "Tanishq"],
    "Department" : ["IT Support", "Management", "Frontend", "Backend", "Analytics", "IT Support"],
    "Salary" : [30000, 60000, 40000, 40000, 50000, 30000],
    "City" : ["Pune", "Kothrod", "Nanded", "Pune", "Kokan", "Pune"]
}

df = pd.DataFrame(Employee)
```

```python
# 1. Display first 3 rows. 
print(df.head(3))
```

```python
# 2. Display last 2 rows.
print(df.tail(2))
```

```python
# 3. Display dataset shape. 
print(df.shape)
```

```python
# 4. Display column names.
print(df.columns)
```

```python
# 5. Display employee records whose salary is greater than 40000.
print(df[df["Salary"] >= 40000])
```

```python
# 6. Display employees from Pune. 
print(df[df["City"] == "Pune"])
```

```python
# Q6. Write a program to display employees whose salary is greater than 50000.
print(df[df["Salary"] > 50000])
```
