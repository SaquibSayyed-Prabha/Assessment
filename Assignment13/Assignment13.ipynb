```python
# Create an Employee DataFrame containing: 
# - Name 
# - Department 
# - Salary 

# Perform the following operations: 
# 1. Sort salary in ascending order. 
# 2. Sort salary in descending order. 
# 3. Display unique departments. 
# 4. Count employees in each department. 
# 5. Find average salary of each department. 
# 6. Find maximum salary of each department. 
```

```python
# Create an Employee DataFrame containing: 
# - Name 
# - Department 
# - Salary 
import pandas as pd

Employee = {
    "Name": ["Saquib", "Atharva", "Jitendra", "Mansi", "Aaryan", "Suraj"],
    "Department": ["HR", "IT", "HR", "Finance", "IT", "Finance"],
    "Salary": [35000, 60000, 40000, 55000, 70000, 50000]
}

df = pd.DataFrame(Employee)

print("Employee DataFrame:")
print(df)
```

```python
# 1. Sort salary in ascending order
print(df.sort_values(by="Salary"))
```

```python
# 2. Sort salary in descending order
print(df.sort_values(by="Salary", ascending=False))
```

```python
# 3. Display unique departments
print(df["Department"].unique())
```

```python
# 4. Count employees in each department
print(df["Department"].value_counts())
```

```python
# 5. Find average salary of each department
print(df.groupby("Department")["Salary"].mean())
```

```python
# 6. Find maximum salary of each department
print(df.groupby("Department")["Salary"].max())
```
