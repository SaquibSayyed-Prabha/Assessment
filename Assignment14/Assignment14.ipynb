```python
# Create a Customer Dataset containing: 
# - Age 
# - Salary 
# - Gender 
# - Purchased 
import pandas as pd

data = {
    "Age": [25, 32, 47, 51, 23, 40, 36, 29],
    "Salary": [50000, 60000, 80000, 90000, 45000, 75000, 65000, 55000],
    "Gender": ["Male", "Female", "Male", "Female", "Female", "Male", "Female", "Male"],
    "Purchased": ["No", "Yes", "Yes", "Yes", "No", "Yes", "No", "No"]
}

df = pd.DataFrame(data)
```

```python
# 1. Display first 5 rows
print(df.head())
```

```python
# 2. Display dataset information
df.info()
```

```python
# 3. Display statistical summary
print(df.describe())
```

```python
# 4. Encode Gender column using map()
df["Gender"] = df["Gender"].map({"Male": 0, "Female": 1})

print("\nDataset after Encoding Gender:")
print(df)
```

```python
# 5. Find correlation between Age and Salary
print(df["Age"].corr(df["Salary"]))
```

```python
# 6. Select features and target columns
features = df[["Age", "Salary", "Gender"]]
print("Features:")
print(features)
print()

target = df["Purchased"]
print("Target:")
print(target)
```
